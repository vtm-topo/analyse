# Checking if some XXX remains

While working on the file transcription, I sometimes used XXX to mark a field where I wasn't sure. I came back to it later to correct it but it is possible that I forgot some. I want to check if that is the case. Did I forgot some xxx? How many?

```python
from lxml import etree

tree = etree.parse("../CHABLE/data/AC Bagnes R46 0052.xml")
root = tree.getroot()
```

we can try to loop through the content of our TextLine

```python
for t in root.iter('{*}TextLine'):
    for String in list(t):
        try:
            if String.attrib['CONTENT'] == "Article":
                print('SHIT')
        except:
            continue
```

Then try to use regex inside with re

```python
import re

for t in root.iter('{*}TextLine'):
    for String in list(t):
        try:
            x = String.attrib['CONTENT']
            if re.search('Article|oranges', x):
                print('Article or oranges in string')
            else:
                print('Neither fruit is in string')
        except:
            continue
```

Now we need to try with a proper regex rules. I will modify a file to contain some xx and XXX

```python
tree = etree.parse("AC Bagnes R46 0201.xml")
root = tree.getroot()
```

```python
y = 0

for t in root.iter('{*}TextLine'):
    for String in list(t):
        try:
            x = String.attrib['CONTENT']
            
            if re.search('x{2,}|X{2,}', x):
                y = y+1
                print('XXXX found OMG!')
                print(x)
            else:
                print('Nope')
        except:
            continue
            
print(y)
            
```

our regex pattern is working fine. It matches with 2x but not with one. Perfect. We can write a script for all files now.
